```python
from meteostat import Monthly, Point
from datetime import datetime

# Locations
chicago = Point(41.8781, -87.6298)
rome = Point(41.9028, 12.4964)

start = datetime(2000, 1, 1)
end = datetime(2025, 12, 31)

chi_monthly = Monthly(chicago, start, end).fetch()
rome_monthly = Monthly(rome, start, end).fetch()

print(chi_monthly.head())
print(rome_monthly.head())
```

```python
from sklearn.preprocessing import StandardScaler
from ripser import ripser
from persim import plot_diagrams
```

```python
def fetch_meteostat_monthly(lat, lon, start, end, label="city"):
    loc = Point(lat, lon)
    df = Monthly(loc, start, end).fetch()

    if df.empty:
        raise ValueError(f"[ERROR] No data returned for {label} (lat={lat}, lon={lon}).")

    if "tavg" not in df.columns:
        raise ValueError(f"[ERROR] {label} missing 'tavg' column. Available: {list(df.columns)}")

    df = df[["tavg"]].dropna()
    df.index = pd.to_datetime(df.index)
    return df


def time_delay_embedding(series, m=3, tau=1):
    x = np.asarray(series, dtype=float)
    n = len(x)
    max_idx = n - (m - 1) * tau
    if max_idx <= 2:
        raise ValueError("[ERROR] Series too short for chosen m and tau.")
    return np.column_stack([x[i:i+max_idx] for i in range(0, m*tau, tau)])


def sliding_window_embedding(series, window=12, step=1):
    x = np.asarray(series, dtype=float)
    n = len(x)
    if n < window + 1:
        raise ValueError("[ERROR] Series too short for chosen window.")
    return np.array([x[i:i+window] for i in range(0, n - window + 1, step)])


def standardize_1d(x):
    return StandardScaler().fit_transform(np.asarray(x).reshape(-1, 1)).ravel()


def standardize_cloud(X):
    return StandardScaler().fit_transform(X)


def run_ripser(X, maxdim=2):
    return ripser(X, maxdim=maxdim)["dgms"]


def persistence_table(diagrams, dim=1):
    """
    Returns a DataFrame with birth/death/persistence for a chosen homology dimension.
    Sorted by persistence descending.
    """
    if len(diagrams) <= dim:
        return pd.DataFrame(columns=["birth", "death", "persistence"])

    D = diagrams[dim]
    if D.size == 0:
        return pd.DataFrame(columns=["birth", "death", "persistence"])

    # Keep only finite deaths
    finite = D[np.isfinite(D[:, 1])]
    if finite.size == 0:
        return pd.DataFrame(columns=["birth", "death", "persistence"])

    births = finite[:, 0]
    deaths = finite[:, 1]
    pers = deaths - births

    out = pd.DataFrame({"birth": births, "death": deaths, "persistence": pers})
    out = out.sort_values("persistence", ascending=False).reset_index(drop=True)
    return out


def print_block(title, lines=None):
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)
    if lines:
        for line in lines:
            print(line)


def summarize_H1(diagrams, topk=5):
    tbl = persistence_table(diagrams, dim=1)
    summary = {
        "H1_count": int(len(tbl)),
        "H1_total_persistence": float(tbl["persistence"].sum()) if len(tbl) else 0.0,
        "H1_max_persistence": float(tbl["persistence"].max()) if len(tbl) else 0.0
    }
    return summary, tbl.head(topk)


def plot_3d_delay_cloud(X, city_label):
    """
    X is delay embedding with columns = [T(t), T(t+1), T(t+2)] in standardized units.
    """
    if X.shape[1] < 3:
        return

    fig = plt.figure(figsize=(7, 6))
    ax = fig.add_subplot(111, projection="3d")
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], s=12)

    ax.set_title(f"{city_label}: Time-Delay Embedding of Monthly Temp (m=3, τ=1)")
    ax.set_xlabel("Temperature at t")
    ax.set_ylabel("Temperature at t+1")
    ax.set_zlabel("Temperature at t+2")

    plt.tight_layout()
    plt.show()


def plot_pd(diagrams, title):
    plt.figure(figsize=(6, 5))
    plot_diagrams(diagrams, show=False)
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
start = datetime(2000, 1, 1)
end   = datetime(2025, 12, 31)

CHICAGO = (41.8781, -87.6298)
ROME    = (41.9028,  12.4964)

m, tau = 3, 1
window, step = 12, 1
```

```python
print_block("1) Fetching Meteostat Monthly Data")

chi = fetch_meteostat_monthly(*CHICAGO, start, end, label="Chicago")
rome = fetch_meteostat_monthly(*ROME, start, end, label="Rome")

common_months = chi.index.intersection(rome.index)
chi = chi.loc[common_months].copy()
rome = rome.loc[common_months].copy()

print(f"Chicago months: {len(chi)} | Rome months: {len(rome)}")
print(f"Common range: {common_months.min().date()} to {common_months.max().date()}")


# Quick time series plot
plt.figure(figsize=(10, 5))
plt.plot(chi.index, chi["tavg"], label="Chicago tavg (°C)")
plt.plot(rome.index, rome["tavg"], label="Rome tavg (°C)")
plt.title("Monthly Average Temperature (Meteostat)")
plt.xlabel("Date")
plt.ylabel("tavg (°C)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
chi_t = standardize_1d(chi["tavg"].values)
rome_t = standardize_1d(rome["tavg"].values)
```

```python
print_block("2) TDA A — Delay Embedding (m=3, τ=1)")

X_chi_delay = time_delay_embedding(chi_t, m=m, tau=tau)
X_rome_delay = time_delay_embedding(rome_t, m=m, tau=tau)

# Standardize point clouds (recommended for stability)
X_chi_delay = standardize_cloud(X_chi_delay)
X_rome_delay = standardize_cloud(X_rome_delay)

plot_3d_delay_cloud(X_chi_delay, "Chicago")
plot_3d_delay_cloud(X_rome_delay, "Rome")

dgms_chi_delay = run_ripser(X_chi_delay, maxdim=2)
dgms_rome_delay = run_ripser(X_rome_delay, maxdim=2)

plot_pd(dgms_chi_delay, "Chicago Persistence Diagrams (Delay Embedding)")
plot_pd(dgms_rome_delay, "Rome Persistence Diagrams (Delay Embedding)")

chi_sum_delay, chi_top_delay = summarize_H1(dgms_chi_delay, topk=5)
rome_sum_delay, rome_top_delay = summarize_H1(dgms_rome_delay, topk=5)

print_block("Delay Embedding — Chicago H1 Summary", [
    f"H1 count            : {chi_sum_delay['H1_count']}",
    f"H1 total persistence: {chi_sum_delay['H1_total_persistence']:.4f}",
    f"H1 max persistence  : {chi_sum_delay['H1_max_persistence']:.4f}",
    "Top H1 features (birth, death, persistence):"
])
print(chi_top_delay.to_string(index=False, float_format=lambda x: f"{x:.4f}"))

print_block("Delay Embedding — Rome H1 Summary", [
    f"H1 count            : {rome_sum_delay['H1_count']}",
    f"H1 total persistence: {rome_sum_delay['H1_total_persistence']:.4f}",
    f"H1 max persistence  : {rome_sum_delay['H1_max_persistence']:.4f}",
    "Top H1 features (birth, death, persistence):"
])
print(rome_top_delay.to_string(index=False, float_format=lambda x: f"{x:.4f}"))
```

```python
print_block("3) TDA B — 12-Month Sliding Window Embedding")

X_chi_win = sliding_window_embedding(chi_t, window=window, step=step)
X_rome_win = sliding_window_embedding(rome_t, window=window, step=step)

X_chi_win = standardize_cloud(X_chi_win)
X_rome_win = standardize_cloud(X_rome_win)

dgms_chi_win = run_ripser(X_chi_win, maxdim=2)
dgms_rome_win = run_ripser(X_rome_win, maxdim=2)

plot_pd(dgms_chi_win, "Chicago Persistence Diagrams (12-Month Sliding Window)")
plot_pd(dgms_rome_win, "Rome Persistence Diagrams (12-Month Sliding Window)")

chi_sum_win, chi_top_win = summarize_H1(dgms_chi_win, topk=5)
rome_sum_win, rome_top_win = summarize_H1(dgms_rome_win, topk=5)

print_block("Sliding Window — Chicago H1 Summary", [
    f"H1 count            : {chi_sum_win['H1_count']}",
    f"H1 total persistence: {chi_sum_win['H1_total_persistence']:.4f}",
    f"H1 max persistence  : {chi_sum_win['H1_max_persistence']:.4f}",
    "Top H1 features (birth, death, persistence):"
])
print(chi_top_win.to_string(index=False, float_format=lambda x: f"{x:.4f}"))

print_block("Sliding Window — Rome H1 Summary", [
    f"H1 count            : {rome_sum_win['H1_count']}",
    f"H1 total persistence: {rome_sum_win['H1_total_persistence']:.4f}",
    f"H1 max persistence  : {rome_sum_win['H1_max_persistence']:.4f}",
    "Top H1 features (birth, death, persistence):"
])
print(rome_top_win.to_string(index=False, float_format=lambda x: f"{x:.4f}"))
```

```python
comparison = pd.DataFrame([
    {"City": "Chicago", "Embedding": "Delay (m=3, τ=1)", **chi_sum_delay},
    {"City": "Rome",    "Embedding": "Delay (m=3, τ=1)", **rome_sum_delay},
    {"City": "Chicago", "Embedding": "Sliding Window (12)", **chi_sum_win},
    {"City": "Rome",    "Embedding": "Sliding Window (12)", **rome_sum_win},
])

print_block("4) Summary Table (copy/paste into report)")
print(comparison.to_string(index=False, float_format=lambda x: f"{x:.4f}"))
```
